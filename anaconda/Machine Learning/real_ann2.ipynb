```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def linear(x):
    return x

def linear_derivative(x):
    return np.ones_like(x)

def mean_squared_error(y_true, y_pred):
    return np.mean((y_pred - y_true) ** 2)

def initialize_parameters(input_nodes, hidden_nodes, output_nodes):
    np.random.seed(42)
    w1 = np.random.randn(input_nodes, hidden_nodes)
    b1 = np.zeros((1, hidden_nodes))
    w2 = np.random.randn(hidden_nodes, output_nodes)
    b2 = np.zeros((1, output_nodes))
    return w1, b1, w2, b2

def forward_propagation(X, w1, b1, w2, b2):
    hidden_output = relu(np.dot(X, w1) + b1)
    final_output = linear(np.dot(hidden_output, w2) + b2)
    return hidden_output, final_output

def backward_propagation(X, y_true, hidden_output, final_output, w2):
    final_error = final_output - y_true
    final_delta = final_error * linear_derivative(final_output)

    hidden_error = np.dot(final_delta, w2.T)
    hidden_delta = hidden_error * relu_derivative(hidden_output)

    return final_delta, hidden_delta

def update_parameters(X, hidden_output, final_delta, hidden_delta, w1, b1, w2, b2, learning_rate):
    w2 -= learning_rate * np.dot(hidden_output.T, final_delta)
    b2 -= learning_rate * np.sum(final_delta, axis=0, keepdims=True)

    w1 -= learning_rate * np.dot(X.T, hidden_delta)
    b1 -= learning_rate * np.sum(hidden_delta, axis=0, keepdims=True)

def train(X, y, hidden_nodes, learning_rate, num_epochs):
    input_nodes = X.shape[1]
    output_nodes = y.shape[1]

    w1, b1, w2, b2 = initialize_parameters(input_nodes, hidden_nodes, output_nodes)

    lt = []
    for epoch in range(num_epochs):
        hidden_output, final_output = forward_propagation(X, w1, b1, w2, b2)

        error = mean_squared_error(y, final_output)
        lt.append(error)

        final_delta, hidden_delta = backward_propagation(X, y, hidden_output, final_output, w2)

        update_parameters(X, hidden_output, final_delta, hidden_delta, w1, b1, w2, b2, learning_rate)

    return w1, b1, w2, b2, lt

# Example usage:
# Assuming you have your data X and target y ready
# Make sure the shapes of X and y are compatible
hidden_nodes = 10
learning_rate = 0.01
num_epochs = 1000
trained_w1, trained_b1, trained_w2, trained_b2, loss_values = train(X, y, hidden_nodes, learning_rate, num_epochs)
print(loss_values)
```
