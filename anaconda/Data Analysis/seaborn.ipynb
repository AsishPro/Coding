### Seaborn
* based on matplotlib
* seaborn is a visualization library in python for statistical plotting 
* Designed to work with dataframe objects in pandas
* It provides high level interface for drawing attractive and informative statistical graphs
* It contains default attaractive style
* https://seaborn.pydata.org

### Categorical plot

* by default it returns scatter plot 
* sns.catplot()
* data in form of characters

```python
help(sns.catplot)  # by default returns scatter plot
```

```python
sns.catplot(x="species",y="sepal_length",data=iris)
```

### Categorical Distribution data
* boxplot

```python
sns.catplot(x="species",y="sepal_length",data=iris,kind="box")
```

```python
sns.catplot(data=iris,kind="box",orient='h')
```

### violin plot
* similar to box plot 

```python

sns.catplot(x="species",y="sepal_width",data=iris,kind="violin")
```

### Joint plot
* combination of two plots , By default it have scatter plot
* sns.jointplot()

```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
sns.jointplot(data=iris)
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x='sepal_width',y='petal_length',kind="hex",data=iris,color='red')
```

```python
sns.jointplot(x='sepal_width',y='petal_length',kind="kde",data=iris,color='red')
```

```python
sns.jointplot(x='sepal_width',y='petal_length',data=iris,color='red')
```

```python
sns.jointplot(x='sepal_width',y='petal_length',kind="hex",data=iris,color='red')
```

### Pairplot 

* It displays multiple graphs at a time in a single graph
* pairwize relationship
* By default it return scatterplot
    * sns.pairplot()
* takes numerical data

```python
sns.pairplot(iris)
```

```python
iris
```

```python
help(sns.pairplot)
```

```python
sns.pairplot(iris,hue="species",palette="rocket")
```

```python
sns.pairplot(iris,hue="species",markers=['o','D','s'])
```

```python
sns.pairplot(iris,vars=['sepal_length','sepal_width'])
```

### Heat map
* It is used for correlation between two columns in dataset.

```python
t=sns.load_dataset("tips")
t
```

```python
t.corr()
```

```python
help(t.corr())
```

```python
sns.heatmap(t.corr(),annot=True)
```

```python
sns.heatmap(t.corr(),center=0)
```

```python
sns.heatmap(t.corr(),center=0,linewidth=10,linecolor="black")
```

```python
sns.heatmap(t.corr(),cmap="Blues")
```
